### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url , sep = '|')
```

### Step 4. Discover what is the mean age per occupation

```python
users.head(20)
```

```python
users.groupby('occupation')['age'].mean()
```

```python
users.head(20)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users['is_male'] = (users['gender'] == 'M').astype(int)
grouped_data = users.groupby('occupation')['is_male'].mean()
grouped_data.sort_values()
```

```python
# using labmda:
users.groupby('occupation')['gender'].apply(lambda x: (x == 'M').mean()).sort_values()
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation')['age'].agg(['min', 'max']) 
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
occupation_gender_MeanAge = users.groupby(['occupation', 'gender'])['age'].mean().unstack('gender')
occupation_gender_MeanAge.fillna(0, inplace=True)
occupation_gender_MeanAge.astype(int)
```

### Step 8.  For each occupation present the percentage of women and men

```python
users['male_percentage'] = users.groupby('occupation')['gender'].transform(lambda x: (x == 'M').mean() * 100).astype(int)
users['female_percentage'] = 100 - users['male_percentage']
users.groupby('occupation')[['male_percentage', 'female_percentage']].mean().astype(int)
```
